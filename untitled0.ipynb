# Cosmolopy

```python
!pip install healpy
!pip install cosmolopy


import numpy as np
import matplotlib.pyplot as plt
import healpy as hp
import cosmolopy.distance as cd

n_points = 10000
nside = 32

# Cosmological model
cosmo = {'omega_M_0': 0.3, 'omega_lambda_0': 0.7, 'omega_k_0': 0.0, 'h': 0.7}
z_max = 2.0

# Inverse transform sampling for redshifts
z_array = np.linspace(0, z_max, 1000)
dV_dz = cd.comoving_volume_element(z_array, **cosmo)
CDF = np.cumsum(dV_dz)
CDF = CDF / CDF[-1]

u = np.random.uniform(0, 1, n_points)
z_samples = np.interp(u, CDF, z_array)

# Sphere sampling
x = np.random.randn(n_points)
y = np.random.randn(n_points)
z = np.random.randn(n_points)

r = np.sqrt(x**2 + y**2 + z**2)
x, y, z = x/r, y/r, z/r

# Convert to healpy angles
theta = np.arccos(z)
phi = np.arctan2(y, x) % (2 * np.pi)

# Bin into healpy pixels
npix = hp.nside2npix(nside)
pixel_indices = hp.ang2pix(nside, theta, phi)

# Compute flux and sum per pixel
L = 1.0
D_L = cd.luminosity_distance(z_samples, **cosmo)
F = L / (4 * np.pi * D_L**2)

flux_map = np.bincount(pixel_indices, weights=F, minlength=npix)

# Plot
hp.mollview(flux_map, title="Total Flux per Healpy Pixel", unit="Flux (arb. units)", cmap="inferno")
hp.graticule()
plt.show()

```

# Flux vs redshift

```python
!pip install healpy
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp
from astropy.cosmology import FlatLambdaCDM

#Settings
cosmo = FlatLambdaCDM(H0=70, Om0=0.3)
N = 200000
nside = 32
npix = hp.nside2npix(nside)

# Sample galaxy redshifts
z_array = np.linspace(0, 1.0, 10000)
cdf = np.cumsum(cosmo.differential_comoving_volume(z_array).value)
cdf /= cdf[-1]
z_samples = np.interp(np.random.rand(N), cdf, z_array)

# rann() code
x = np.random.randn(N)
y = np.random.randn(N)
z = np.random.randn(N)

r = np.sqrt(x**2 + y**2 + z**2)  # length of each vector
x, y, z = x/r, y/r, z/r          # normalise to unit sphere

#  Convert positions to healpy pixels
theta = np.arccos(z)                   # polar angle
phi   = np.arctan2(y, x) % (2*np.pi)  # azimuthal angle
pixels = hp.ang2pix(nside, theta, phi)

# Calculate flux for each galaxy
D_L = cosmo.luminosity_distance(z_samples).value
flux = 1.0 / (4 * np.pi * D_L**2)

# Average flux in each pixel
flux_per_pixel  = np.bincount(pixels, weights=flux, minlength=npix)
count_per_pixel = np.bincount(pixels, minlength=npix).astype(float)
avg_flux = np.where(count_per_pixel > 0, flux_per_pixel / count_per_pixel, hp.UNSEEN)

log_flux = np.where(avg_flux != hp.UNSEEN, np.log10(avg_flux), hp.UNSEEN)

# sky map
hp.mollview(log_flux, title="Average Flux per Pixel", unit="log10(Flux)",
            cmap="inferno",
            min=np.percentile(log_flux[log_flux != hp.UNSEEN], 5),
            max=np.percentile(log_flux[log_flux != hp.UNSEEN], 95))
hp.graticule()
plt.show()

# sanity check Flux vs redshift
plt.figure(figsize=(8, 5))
plt.scatter(z_samples, flux, s=1, alpha=0.3)
plt.xlabel("Redshift z")
plt.ylabel("Flux")
plt.title("Flux vs Redshift")
plt.yscale("log")
plt.show()
```

# Dipole with Likelihood

```python
!pip install emcee
!pip install Corner
!pip install healpy
import numpy as np
import healpy as hp
import emcee
import corner
import matplotlib.pyplot as plt

# 1. SIMULATE DATA
nside = 32
npix = hp.nside2npix(nside)
beta_true = 0.1
D_true = 2 * beta_true

# Create random sky
n_points = 500000
x = np.random.randn(n_points)
y = np.random.randn(n_points)
z = np.random.randn(n_points)
r = np.sqrt(x**2 + y**2 + z**2)
x, y, z = x/r, y/r, z/r

# Relativistic Boost (z-direction)
z_new = (z + beta_true) / (1 + beta_true * z)
gamma = 1 / np.sqrt(1 - beta_true**2)
factor = 1 / (gamma * (1 + beta_true * z))
x_new, y_new = x * factor, y * factor

# Bin into map
pix = hp.ang2pix(nside, np.arccos(z_new), np.arctan2(y_new, x_new))
counts = np.bincount(pix, minlength=npix).astype(float)

# Pre-calculate Pixel Unit Vectors (Fixed)
th_p, ph_p = hp.pix2ang(nside, np.arange(npix))
px = np.sin(th_p) * np.cos(ph_p)
py = np.sin(th_p) * np.sin(ph_p)
pz = np.cos(th_p)

# 2. LEAST SQUARES
# Solve: Counts = A + dx*X + dy*Y + dz*Z
X_mat = np.vstack([np.ones(npix), px, py, pz]).T
beta_ls, _, _, _ = np.linalg.lstsq(X_mat, counts, rcond=None)

A_start = beta_ls[0]
dx, dy, dz = beta_ls[1:] / A_start
D_start = np.sqrt(dx**2 + dy**2 + dz**2)
theta_start = np.arccos(dz / D_start)
phi_start = np.arctan2(dy, dx) % (2 * np.pi)

print(f"Least Squares Guess: D={D_start:.3f}, theta={np.degrees(theta_start):.1f}")

# 3. LIKELIHOOD
def log_prob(p):
    A, D, theta, phi = p

    # Priors
    if not (0 < A < 50 and 0 <= D < 1 and 0 < theta < np.pi and 0 <= phi < 2*np.pi):
        return -np.inf

    # Convert MCMC guess (angles) to unit vector u
    ux = np.sin(theta) * np.cos(phi)
    uy = np.sin(theta) * np.sin(phi)
    uz = np.cos(theta)

    # Dot product u · p
    model = A * (1 + D * (ux*px + uy*py + uz*pz))

    if np.any(model <= 0): return -np.inf

    # Poisson + Volume Prior
    return np.sum(counts * np.log(model) - model) + np.log(np.sin(theta))

# 4. RUN MCMC
ndim, nwalkers = 4, 12
# Start walkers in a tiny ball around the Least Squares result
initial_pos = [A_start, D_start, theta_start, phi_start] + 1e-4 * np.random.randn(nwalkers, ndim)

sampler = emcee.EnsembleSampler(nwalkers, ndim, log_prob)
print("Sampling...")
sampler.run_mcmc(initial_pos, 1000, progress=True)

# 5. CORNER PLOT
samples = sampler.get_chain(discard=200, flat=True)
# Convert angles to degrees for readability
samples_deg = samples.copy()
samples_deg[:, 2:] = np.degrees(samples_deg[:, 2:])

corner.corner(samples_deg, labels=["A", "D", "theta (deg)", "phi (deg)"], show_titles=True)
plt.show()
```

# Dipole Sky Map

```python
!pip install healpy
!pip install emcee
!pip install Corner

import numpy as np
import healpy as hp
import emcee
import corner
import matplotlib.pyplot as plt

# -------------------------------
# 1. simulate boosted catalogue
# -------------------------------
np.random.seed(42)

n_points = 500_000
nside = 64
npix = hp.nside2npix(nside)

beta_true = 0.1
D_true = 2 * beta_true

# isotropic sphere
x, y, z = np.random.randn(3, n_points)
r = np.sqrt(x**2 + y**2 + z**2)
x, y, z = x/r, y/r, z/r

# relativistic boost (z-direction)
gamma = 1 / np.sqrt(1 - beta_true**2)
z_new = (z + beta_true) / (1 + beta_true*z)
fac = 1 / (gamma * (1 + beta_true*z))
x_new = x * fac
y_new = y * fac

# healpix binning
pix = hp.ang2pix(
    nside,
    np.arccos(np.clip(z_new, -1, 1)),
    np.arctan2(y_new, x_new)
)
counts = np.bincount(pix, minlength=npix).astype(float)

# pixel directions
theta_pix, phi_pix = hp.pix2ang(nside, np.arange(npix))
x_pix = np.sin(theta_pix)*np.cos(phi_pix)
y_pix = np.sin(theta_pix)*np.sin(phi_pix)
z_pix = np.cos(theta_pix)

# -------------------------------
# 2. initial guess
# -------------------------------
X = np.vstack([np.ones(npix), x_pix, y_pix, z_pix]).T
params, _, _, _ = np.linalg.lstsq(X, counts, rcond=None)

A0 = params[0]
dx0, dy0, dz0 = params[1:] / A0
D0 = np.sqrt(dx0**2 + dy0**2 + dz0**2)

theta0 = np.arccos(np.clip(dz0 / D0, -1, 1))
phi0 = np.arctan2(dy0, dx0) % (2*np.pi)

# -------------------------------
# 3. likelihood
# -------------------------------
def log_prob(p):
    A, D, theta, phi = p

    if A <= 0 or D < 0 or D >= 2:
        return -np.inf
    if theta <= 0 or theta >= np.pi:
        return -np.inf
    if phi < 0 or phi >= 2*np.pi:
        return -np.inf

    ux = np.sin(theta)*np.cos(phi)
    uy = np.sin(theta)*np.sin(phi)
    uz = np.cos(theta)

    lam = A * (1 + D*(ux*x_pix + uy*y_pix + uz*z_pix))
    if np.any(lam <= 0):
        return -np.inf

    return np.sum(counts*np.log(lam) - lam) + np.log(np.sin(theta))

# -------------------------------
# 4. MCMC
# -------------------------------
ndim, nwalkers = 4, 32
start = np.array([A0, D0, theta0, phi0])
pos = start + 1e-3*np.random.randn(nwalkers, ndim)
pos[:,3] %= 2*np.pi

sampler = emcee.EnsembleSampler(nwalkers, ndim, log_prob)

print("Burn-in...")
pos, _, _ = sampler.run_mcmc(pos, 500, progress=True)
sampler.reset()

print("Production...")
sampler.run_mcmc(pos, 1500, progress=True)

samples = sampler.get_chain(discard=200, thin=10, flat=True)

A_fit, D_fit, theta_fit, phi_fit = np.median(samples, axis=0)
print("Recovered beta =", D_fit/2)

# -------------------------------
# 5. corner plot
# -------------------------------
samples_plot = samples.copy()
samples_plot[:,2] = np.degrees(samples_plot[:,2])
samples_plot[:,3] = np.degrees(samples_plot[:,3])

corner.corner(
    samples_plot,
    labels=["A","D","theta (deg)","phi (deg)"],
    quantiles=[0.16,0.5,0.84],
    show_titles=True
)

plt.show()
# -------------------------------
# 6. build sky posterior (CENTERED)
# -------------------------------

theta_s = samples[:,2]
phi_s   = samples[:,3]

# convert samples to vectors
ux = np.sin(theta_s)*np.cos(phi_s)
uy = np.sin(theta_s)*np.sin(phi_s)
uz = np.cos(theta_s)
vecs = np.vstack([ux, uy, uz]).T

# --- find median direction (to centre map) ---
theta_med = np.median(theta_s)
phi_med   = np.median(phi_s)

v0 = np.array([
    np.sin(theta_med)*np.cos(phi_med),
    np.sin(theta_med)*np.sin(phi_med),
    np.cos(theta_med)
])

# target direction = centre of mollview (x-axis)
target = np.array([1, 0, 0])

# compute rotation matrix
axis = np.cross(v0, target)
norm = np.linalg.norm(axis)

if norm < 1e-10:
    R = np.eye(3)
else:
    axis /= norm
    angle = np.arccos(np.clip(np.dot(v0, target), -1, 1))

    K = np.array([
        [0, -axis[2], axis[1]],
        [axis[2], 0, -axis[0]],
        [-axis[1], axis[0], 0]
    ])

    R = np.eye(3) + np.sin(angle)*K + (1-np.cos(angle))*(K @ K)

# rotate all sample vectors
vecs_rot = vecs @ R.T

# -------------------------------
# build map
# -------------------------------
sky_map = np.zeros(npix)

for v in vecs_rot:
    dot = v[0]*x_pix + v[1]*y_pix + v[2]*z_pix
    sky_map += np.exp(15 * dot)

sky_map /= np.max(sky_map)

# -------------------------------
# 7. plot mollview (BETTER VISUALS)
# -------------------------------

# log stretch for better contrast
plot_map = np.log10(sky_map + 1e-6)

hp.mollview(
    plot_map,
    title="Dipole Direction Posterior ",
    cmap="inferno",          # much better contrast
    unit="log density",
    min=plot_map.max() - 3   # enhances dynamic range
)

# grid lines
hp.graticule(dpar=30, dmer=30, verbose=False)

# add manual degree labels
for lon in range(-150, 180, 30):
    hp.projtext(
        lon, 0,
        f"{lon}°",
        lonlat=True,
        fontsize=8,
        color="white",
        ha="center"
    )

for lat in [-60, -30, 0, 30, 60]:
    hp.projtext(
        0, lat,
        f"{lat}°",
        lonlat=True,
        fontsize=8,
        color="white",
        va="center"
    )


plt.show()
```

```python
!pip install healpy
!pip install emcee
!pip install matplotlib
!pip install healpy
!pip install emcee
!pop install corner

import numpy as np
import healpy as hp
import emcee
import corner
import matplotlib.pyplot as plt
from healpy.projector import MollweideProj

# simulate boosted sky

n_points = 500_000
nside    = 64
npix     = hp.nside2npix(nside)
beta     = 0.1

# generate random points on a unit sphere
x, y, z = np.random.randn(3, n_points)
r = np.sqrt(x**2 + y**2 + z**2)
x, y, z = x/r, y/r, z/r

# boost in z direction using relativistic aberration
gamma = 1 / np.sqrt(1 - beta**2)
z_new = (z + beta) / (1 + beta*z)
fac   = 1 / (gamma * (1 + beta*z))
x_new, y_new = x*fac, y*fac

# bin directions into healpix pixels
# convert the boosted 3D directions to (theta, phi) angles then to pixel indices.
# clip keeps z_new in [-1,1] to avoid arccos blowing up
pix    = hp.ang2pix(nside, np.arccos(np.clip(z_new, -1, 1)), np.arctan2(y_new, x_new))
#observed data
counts = np.bincount(pix, minlength=npix).astype(float)

# unit vector for each pixel centre
theta_pix, phi_pix = hp.pix2ang(nside, np.arange(npix))
x_pix = np.sin(theta_pix)*np.cos(phi_pix)
y_pix = np.sin(theta_pix)*np.sin(phi_pix)
z_pix = np.cos(theta_pix)

# initial guess via least squares, put walkers near

# fit counts = A(1 + D*nx) using columns [1, x, y, z]
X = np.vstack([np.ones(npix), x_pix, y_pix, z_pix]).T
params, _, _, _ = np.linalg.lstsq(X, counts, rcond=None)

A0         = params[0]
dx, dy, dz = params[1:] / A0
D0         = np.sqrt(dx**2 + dy**2 + dz**2)
theta0     = np.arccos(np.clip(dz / D0, -1, 1))
phi0       = np.arctan2(dy, dx) % (2*np.pi)

# log probability with two angles and jacobian

def log_prob(p):
    A, D, theta, phi = p
#Our parameters
    if A <= 0 or not (0 < D < 2):  return -np.inf
    if not (0 < theta < np.pi):    return -np.inf
    if not (0 <= phi < 2*np.pi):   return -np.inf

    # dipole direction as unit vector
    ux = np.sin(theta)*np.cos(phi)
    uy = np.sin(theta)*np.sin(phi)
    uz = np.cos(theta)

    # expected counts per pixel under dipole model
    lam = A * (1 + D*(ux*x_pix + uy*y_pix + uz*z_pix))
    if np.any(lam <= 0): return -np.inf

    # poisson log likelihood + sin(theta) jacobian for flat prior on sphere
    return np.sum(counts*np.log(lam) - lam) + np.log(np.sin(theta))

# run mcmc

ndim, nwalkers = 4, 32

# start walkers around the least squares guess
pos = np.array([A0, D0, theta0, phi0]) + 1e-3*np.random.randn(nwalkers, ndim)

pos[:, 2] = np.clip(pos[:, 2], 1e-3, np.pi - 1e-3)
pos[:, 3] %= 2*np.pi

#call from every walker
sampler = emcee.EnsembleSampler(nwalkers, ndim, log_prob)

#From emcee
print("burn-in..."); pos, _, _ = sampler.run_mcmc(pos, 500, progress=True); sampler.reset()
print("production..."); sampler.run_mcmc(pos, 1500, progress=True)

samples = sampler.get_chain(discard=200, thin=10, flat=True)
print("recovered beta =", np.median(samples[:, 1]) / 2)

# corner plot
#convert angles to degree for the plot
s = samples.copy()
s[:, 2] = np.degrees(s[:, 2])
s[:, 3] = np.degrees(s[:, 3])

#corner formating
corner.corner(
    s,
    labels=["A", "D", "theta (deg)", "phi (deg)"],
    quantiles=[0.16, 0.5, 0.84],
    show_titles=True,
    smooth=1, smooth1d=1,
    bins=40,
    plot_datapoints=False,
    fill_contours=True,
    title_kwargs={"fontsize": 11},
    label_kwargs={"fontsize": 11},
)
plt.show()

# build sky map centered on mean direction
#convert every MCMC sample direction back into a 3D unit vector
theta_s, phi_s = samples[:, 2], samples[:, 3]
ux = np.sin(theta_s)*np.cos(phi_s)
uy = np.sin(theta_s)*np.sin(phi_s)
uz = np.cos(theta_s)
vecs = np.vstack([ux, uy, uz]).T

# average of all the sample vectors — this is the mean dipole direction
v0   = np.mean(vecs, axis=0); v0 /= np.linalg.norm(v0)
axis = np.cross(v0, [1,0,0]); norm = np.linalg.norm(axis)

#if cros product is 0= {1,0 , 0} or else need to rotate
if norm < 1e-10:
    R = np.eye(3)
else:
    axis /= norm; angle = np.arccos(np.clip(np.dot(v0, [1,0,0]), -1, 1))

    #rodrigues matrix, no trig for each needed
    K = np.array([[0,-axis[2],axis[1]],[axis[2],0,-axis[0]],[-axis[1],axis[0],0]])
    R = np.eye(3) + np.sin(angle)*K + (1-np.cos(angle))*(K@K)

#all samples have been rotated so mean sits in centre
vecs_rot = vecs @ R.T

# Gives cos(angle) between the sample and each pixel
sky_map = np.zeros(npix)
for v in vecs_rot:
    sky_map += np.exp(15*(v[0]*x_pix + v[1]*y_pix + v[2]*z_pix))
sky_map /= sky_map.max()

# credible region levels

# find pixel thresholds enclosing 68% and 95% of total probability
prob_map    = sky_map / sky_map.sum()
sorted_prob = np.sort(prob_map)[::-1]
cum_prob    = np.cumsum(sorted_prob)
level_68    = sorted_prob[np.searchsorted(cum_prob, 0.68)]
level_95    = sorted_prob[np.searchsorted(cum_prob, 0.95)]

# plot sky map with contours

hp.mollview(np.log10(sky_map + 1e-6), title="dipole direction posterior", cmap="inferno", unit="log density")
hp.graticule(dpar=30, dmer=30, verbose=False)

# project grid into  mollweide coordinates so contours align with the map
tt, pp = np.meshgrid(np.linspace(0.01, np.pi-0.01, 300), np.linspace(0, 2*np.pi, 600))
grid   = prob_map[hp.ang2pix(nside, tt, pp)]
xx, yy = MollweideProj().ang2xy(tt.ravel(), pp.ravel())
plt.gca().contour(xx.reshape(tt.shape), yy.reshape(tt.shape), grid,
                  levels=[level_95, level_68], colors=["white", "Wite"], linewidths=1.5)

plt.show()
```
